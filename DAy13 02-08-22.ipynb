```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.arange(1,11)
print(x)

#random data
y=np.random.randint(1,100,10)
print(y)


plt.plot(x,y)
plt.show()
```

```python
x=np.arange(1,11)
print(x)

#random data
y=np.random.randint(1,100,10)
print(y)


plt.plot(x,y,color='red',linewidth=3,marker='*',markersize=10,markeredgecolor='blue')
plt.show()
```

```python
t=np.arange(0,2,0.01)

s=np.sin(2*np.pi*t)

plt.plot(t,s)
plt.show()
```

```python
t=np.arange(0,2,0.01)

s=np.sin(2*np.pi*t)

h=np.cos(2*np.pi*t)

plt.plot(t,s)
plt.plot(t,h)
plt.show()
```

```python
t=np.arange(0,2,0.01)

s=np.sin(2*np.pi*t)

h=np.cos(2*np.pi*t)

plt.plot(t,s,label='sin')
plt.plot(t,h,label='cos')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(13,6))

t=np.arange(0,2,0.01)

s=np.sin(2*np.pi*t)

h=np.cos(2*np.pi*t)

plt.subplot(2,1,1)
plt.plot(t,s,label='sin')

plt.subplot(2,1,2)
plt.plot(t,h,label='cos')
plt.legend()
plt.show()
```

```python
import csv
x=[]
y=[]
with open('data1.csv','r') as f:
    data=csv.reader(f)
    for col in data:
        x.append(col[0])
        y.append(col[1])
        
print('x=',x)
print('y=',y)

plt.plot(x,y)
plt.show()
```
