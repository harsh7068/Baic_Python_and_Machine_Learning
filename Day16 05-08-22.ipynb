```python
google=pd.read_csv('google_stock_price.csv',squeeze=True)
google
```

```python
def classify_per(number):
    if number<300:
        return 'OK'
    elif number>-300 and number <650:
        return 'Satisfactory'
    else:
        return 'Incredible'
```

```python
#apply method
google.apply(classify_per)
```

```python
google.apply(classify_per)[2500:2700]
```
