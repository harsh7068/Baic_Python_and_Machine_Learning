```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
t=np.arange(0,2,0.01)
s=np.sin(2*np.pi*t)

plt.plot(t,s)

plt.show()
```

```python
t=np.arange(0,2,0.01)
s=np.sin(2*np.pi*t)
h=np.cos(2*np.pi*t)
plt.plot(t,s)
plt.plot(t,h)
plt.show()
```

```python
#subplot(row,column,section)
t=np.arange(0,2,0.01)
s=np.sin(2*np.pi*t)
h=np.cos(2*np.pi*t)
plt.subplot(1,2,1)
plt.plot(t,s)


plt.subplot(1,2,2)
plt.plot(t,h)

plt.show()
```

```python
#subplot(row,column,section)
t=np.arange(0,2,0.01)
s=np.sin(2*np.pi*t)
h=np.cos(2*np.pi*t)
plt.subplot(1,2,1)
plt.plot(t,s)
plt.grid()

plt.subplot(1,2,2)
plt.plot(t,h)
plt.grid()
plt.show()
```
