```python
import pandas as pd
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('Position_Salaries.csv')
data
```

```python
real_x=data.iloc[:,1].values
real_x=real_x.reshape(-1,1)
real_x
```

```python
real_y=data.iloc[:,2].values
real_y
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
reg=RandomForestRegressor(n_estimators=400,random_state=0)
#n_estimators ki values jitni jada hogi utna hi tree bnega aur data utna hi accurate hota jaega
```

```python
reg.fit(real_x,real_y)
```

```python
y_pred=reg.predict([[6]])
y_pred
```

```python
x_grid=np.arange(min(real_x),max(real_x),0.01)
x_grid
```

```python
x_grid=x_grid.reshape(len(x_grid),1)
x_grid
```

```python
plt.scatter(real_x,real_y,color='blue')
plt.plot(x_grid,reg.predict(x_grid),color='red')
plt.grid()
plt.show()
```
