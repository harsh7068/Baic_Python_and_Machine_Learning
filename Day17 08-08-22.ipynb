```python
import pandas as pd
```

```python
nba=pd.read_csv('nba.csv')
nba
```

```python
#data cleaning
#dropna()
nba.dropna()


#ek v coloumn me entry missing hai to puri row gyb kr dega
```

```python
#dropna()--->any,all
#by default any call hota hai 
nba.dropna(how='all')
```

```python
#subset
nba.dropna(subset=['Salary'])
```

```python
nba.dropna(subset=['Salary','College'])
```

```python
#fillna()
nba.fillna(0)
```

```python
nba['Salary'].fillna(0,inplace=True)
nba
```

```python
nba['College'].fillna('No college',inplace=True)
nba
```

```python
#sort_values method with data frame
nba.sort_values('Name')
```

```python
nba.sort_values('Name',ascending=False)
```

```python
nba.sort_values('Salary',na_position='first')
```

```python
nba.sort_values(['Team','Salary'])

#team ke according sorting hui kyunki wo phle likha hai to preference jada mila hai usko 
```

```python
nba.sort_values(['Team','Salary'],ascending=[False,False],inplace=True)
nba
```
