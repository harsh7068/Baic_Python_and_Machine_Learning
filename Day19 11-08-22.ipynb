```python
import pandas as pd
```

```python
emp=pd.read_csv('employees.csv',parse_dates=['Start Date','Last Login Time'])
emp['Gender']=emp['Gender'].astype('category')
emp['Senior Management']=emp['Senior Management'].astype('bool')
emp
```

```python
mask = emp['Team']=='Finance'
```

```python
emp[mask]
```

```python
#logical operation ---> AND(&), or(|), not(~)

mask1=emp['Gender']=='Male'
mask2=emp['Team']=='Finance'

emp[mask1 & mask2]
```

```python
emp[mask1 | mask2]
```

```python
emp[~mask1]
```

```python
mask3=emp['Team']=='Legal'
mask4=emp['Team']=='Sales'
mask5=emp['Team']=='Product'

emp[mask3 | mask4 | mask5]
```

```python
emp[(mask3 & mask4) | mask5]
```

```python
mask=emp['Team'].isin(['Legal','Sales','Product'])
emp[mask]
```

```python
emp[emp['Team'].isin(['Legal','Sales','Product'])]
```

```python
#how to check data is null or not null
#isnull()
#notnull()
mask=emp['Team'].isnull()
emp[mask]
```

```python
emp.isnull()
#jahan false hai mtlb entry hai data me
```

```python
mask1=emp['Team'].notnull()
emp[mask1]
```

```python
#between method
emp['Salary'].between(60000,70000)
```

```python
emp[emp['Salary'].between(60000,70000)]
```

```python
emp[emp['Bonus %'].between(2.0 , 5.0)]
```
